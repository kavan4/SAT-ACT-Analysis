# A Strategic Solution to Improve SAT Participation Rates

-  Which regions/states are the strongest and why? 
-  Which regions/states are the weakest and why?

## Problem Statement

Low participation rates (mode of 3%) in the MidWest has lead to downward pressure on revenue and marketshare for the College Board (SAT).

By comparison, the ACT has achieved participation rates of 100% in the MidWest (particularly in states like Minnesota and Wisconsin) by lobbying for statewide graduation requirements. 

A successful solution would be to increase lobbying of state government and provide free resources to educate students and parents to increase brand awarenes.

## Background
The College Board is a not-for-profit organzation best known for adminstering a fee-based standardized test for college admissions known as the SAT. 

In 2011, another standardized test for college admissions, the ACT, surpassed the SAT in terms of total test takers. This prompted the SAT to release a new format in March 2016 to better align with collegiate expectations and legislative objectives.

In order to examine how the SAT can increase marketshare and revenue, I compared participation rates from SAT and ACT datasets for the years 2017 and 2018. 

### Takeaways

We identified the MidWest as a region to focus our efforts - particularly in states such as Ohio, Minnesotta, and Illinois. We observed the largest gaps in participation rates in these regions often because the ACT has statewide commitments. 

Also, we observed that students in the regions tend to have a higher Total Score than other regions due to the fact that students who take the non-mandatory SAT tend to select the test and are better prepared than those students who take the test as a graduation requirement. These are helpful outcomes to use in negotiation.

#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

```python
def std_manual(col):
    return np.sqrt((np.sum((final[col]-np.mean(final[col]))**2)/len(final)))

#Here I hardcode the dataframe "final" into the return function and pass in the column_name: sat_math_2017, act_math_2017, etc.
```

```python
std_manual('sat_math_2018')
```

- Use a **dictionary comprehension** to apply the standard deviation function to each numeric column in the dataframe.  **No loops**  
- Assign the output to variable `sd` as a dictionary where: 
    - Each column name is now a key 
    - That standard deviation of the column is the value 
     
*Example Output :* `{'ACT_Math': 120, 'ACT_Reading': 120, ...}`

#### Define a custom function to subplot histograms

We have data for two tests for two years. We only have composite (and not subtest scores) for the 2018 ACT. We should write a function that will take the names of 2+ columns and subplot histograms. While you can use pandas plotting or Seaborn here, matplotlib gives you greater control over all aspects of your plots.

Here's some starter code:

```python
def subplot_histograms(df, list_of_columns, list_of_titles, list_of_xlabels):
    nrows = int(np.ceil(len(list_of_columns)/2)) 
    fig, ax = plt.subplots(nrows=nrows, ncols=2, figsize=(20, 5*nrows), sharey='all')
    ax = ax.ravel()
    
    if len(list_of_columns) %2 !=0:
        ax[len(list_of_columns)].set_visible(False)
        
    for i, column in enumerate(list_of_columns): # Gives us an index value to get into all our lists
        ax[i].hist(df[column], bins=15) # feel free to add more settings
        ax[i].set_title(list_of_titles[i], fontsize=20)
        ax[i].set_ylabel('Count of States', fontsize=15)
        ax[i].set_xlabel(list_of_xlabels[i], fontsize=15)

    plt.tight_layout()
```
